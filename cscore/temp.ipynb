```python
import pandas as pd
```

```python
data = pd.read_csv('./output_backup/all_csore_createat20250402_120844.csv')
```

```python
data
```

```python
with pd.ExcelWriter('output_by_final_score.xlsx', engine='xlsxwriter') as writer:
    # Group by the 'final_score' column
    for score, group_df in data.groupby('final_score'):
        # Convert the score to string to use as a sheet name
        sheet_name = str(score)
        # Optional: limit sheet name length to 31 characters (Excel limit)
        sheet_name = sheet_name[:31]
        group_df.to_excel(writer, sheet_name=sheet_name, index=False)
```
